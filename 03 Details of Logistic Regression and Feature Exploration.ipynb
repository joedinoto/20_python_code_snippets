## Linear Correlation

You can use $\LaTeX$ in markdown! The expected value $E[ \ ]$ is the mean in the pearson correlation below.

$$
\rho = \dfrac{E[(X - \mu_X)(Y - \mu_Y)]}{\sigma_X\sigma_Y}
$$

We'll be using ANOVA F-tests below because we're dealing with categories and in that context the p-values answer the question:

> "For the samples in the positive class, how likely is it that the average value of this feature is the same as that of samples in the negative class?"

```python
X = df[features_response].iloc[:,:-1].values
y = df[features_response].iloc[:,-1].values
print(X.shape, y.shape)
```

```python
y
```

```python
# I had to invoke
#  from sklearn.feature_selection import f_classif 
# in the header of this doc for this to work

# There are two outputs with this function, the F stat and the p value.
[f_stat, f_p_value] = f_classif(X,y)
```

```python
# this creates a table with three columns 
# the feature (a list of fields), the F stat, and the p value
# then we sort them according to p-value (smallest at the top)

f_test_df = pd.DataFrame({'Feature':features_response[:-1],
                         'F statistic':f_stat,
                         'p value':f_p_value})
f_test_df.sort_values('p value')
```

> We can conclude that `PAY_1` and `LIMIT_BAL` are the most useful features because they have the smallest p values indicating that the average values of these features between the + and - cases are significantly different and can help predict which class a sample belongs to.

```python
# let's pick the top 20% of F-test scores
selector = SelectPercentile(f_classif, percentile = 20)
selector.fit(X,y)
best_feature_ix = selector.get_support()
best_feature_ix
```

```python
# There's got to be a better way to pick out the Trues from my list
# here is the list comprehension method given in the book
features = features_response[:-1]
best_features = [features[counter] for counter in range(len(features)) if best_feature_ix[counter]]
best_features
```

We used an F test however but since we're doing binary classification and comparing means between + and - classes, what we actually did was a t-test which compares the means between two groups of samples to see if the means are statistically significant.

## Sigmoid and log odds function

The sigmoid function is defined as: 

$\text{sigmoid}(x) = \dfrac{1}{1-e^{-x}}$. 

Solving for $x$ we obtain the logit or log odds function 

$x = \ln\left(\dfrac{p}{1-p}\right)$.

```python
# plotting the sigmoid function

# define the sigmoid function
def sigmoid(X):
    Y = 1/(1+np.exp(-X))
    return Y

X_sig = np.linspace(-7,7,141) # domain is values of x from -7 to 7 in 144 evenly spaced intervals
Y_sig = sigmoid(X_sig)

# plotting
plt.plot(X_sig,Y_sig)
plt.yticks(np.linspace(0,1,11)) # y ticks from 0 to 1 in 11 evenly spaced intervals
plt.grid()
plt.title('the Sigmoid function');
```

## create a matrix from a set of  lists using `np.block()`

```python
# this creates a 2x2 matrix from our data above.
X = np.block([[X_1_pos, X_2_pos], [X_1_neg, X_2_neg]]) 
print(X.shape)
print(X[0:3])
```

## create a 'vertial stack' using `np.vstack()`

```python
y = np.vstack( (np.ones((20,1)), np.zeros((20,1)))).reshape(40,) # that's a lot of parenthesis!
print(y[0:5])
print(y[-5:])
```

```python
example_lr = LogisticRegression(solver = 'liblinear')
example_lr.fit(X,y); # the author's use of caps and lowercase for variables is at times frustrating
y_pred = example_lr.predict(X)
positive_indices = [counter for counter in range(len(y_pred)) if y_pred[counter]==1]
negative_indices = [counter for counter in range(len(y_pred)) if y_pred[counter]==0]
positive_indices
```

## plot the classes and predictions on a single scatterplot

```python
plt.scatter(X_1_pos, X_2_pos, color = 'red', marker = 'x')
plt.scatter(X_1_neg, X_2_neg, color = 'blue', marker = 'x')
plt.scatter(X[positive_indices,0], X[positive_indices,1], 
            s = 150, marker = 'o', edgecolors = 'red', facecolors = 'none')
plt.scatter(X[negative_indices,0], X[negative_indices,1], 
            s = 150, marker = 'o', edgecolors = 'blue', facecolors = 'none')
plt.xlabel('$X_1$')
plt.xlabel('$X_2$')
plt.legend(['Positive class', 'Negative class', 'Positive predictions', 'Negative predictions']);
```

## visualizing the decision boundary point

The author is using $\theta_n$ for constants so I'm going with it. 

$X_2 \geq -\dfrac{\theta_1}{\theta_2}X_1 - \dfrac{\theta_0}{\theta_2}$

```python
theta_1 = example_lr.coef_[0][0]
theta_2 = example_lr.coef_[0][1]
theta_0 = example_lr.intercept_
print(theta_1, theta_2, theta_0)
```

```python
X_1_decision_boundary = np.array([0,10])
X_2_decision_boundary = -(theta_1/theta_2)*X_1_decision_boundary - (theta_0/theta_2)
```
